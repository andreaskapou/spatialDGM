# Load trained model and MNIST dataset

```python
# Dataset modification
modify = 1 # rotate or none
z_dim = 2

# Dataset
dm = MnistDataModule(data_dir = osp.join('..', 'data'), dataset_name="MNIST", modify=1)
dm.setup()

# Pyro SpatialVAE model
base_dir = f'../output/PyroVAE_mnist_m{modify}_z{z_dim}/'
pyrovae = load_pyrovae(base_dir, data_dim = (1, 28, 28))
pyrovae = pyrovae.eval()

base_dir = f'../output/PyroVAE_mnist_m{0}_z{z_dim}/'
vae = load_pyrovae(base_dir, data_dim = (1, 28, 28))
vae = vae.eval()
```

# Model predictions

## Training data

```python
plt.rcParams['figure.figsize'] = [12, 4]
viz_num = 12
train_loader = dm.train_dataloader()
viz_idxs = np.linspace(0, len(train_loader.dataset) - 1, viz_num).astype(int)
w, h = plt.rcParams.get('figure.figsize')
fig, axs = plt.subplots(5, len(viz_idxs), figsize=(w * 2, h * 2), facecolor='w')
                        
plot_predictions_pyro(dm_loader=train_loader, viz_idxs=viz_idxs, axs=axs, vae=vae, pyrovae=pyrovae)
```

## Test data

```python
test_loader = dm.val_dataloader()
viz_idxs = np.linspace(0, 100 - 1, viz_num).astype(int)
w, h = plt.rcParams.get('figure.figsize')
fig, axs = plt.subplots(5, len(viz_idxs), figsize=(w * 2, h * 2), facecolor='w')
                        
plot_predictions_pyro(dm_loader=test_loader, viz_idxs=viz_idxs, axs=axs, vae=vae, pyrovae=pyrovae)
```
